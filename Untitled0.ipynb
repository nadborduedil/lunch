```python
from datetime import datetime
from collections import Counter
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt
import statsmodels.api as sm
import re

with open("great_lunch.tsv", "rb") as gls:
    lines = gls.readlines()
    
class Lunch(object):
    def __init__(self):
        self.lower_scale = ""
        self.upper_scale = ""
        self.prompt = ""
        self.ratings = []
        self.people2ratings = {}
        
        
    def make_dict(self):
        for r in self.ratings:
            self.people2ratings[r[2]] = r
        
def read_rating(fields):
    try:
        date = datetime.strptime(fields[0], "%d/%m/%Y %H:%M:%S")
    except:
        return None
    
    return date, int(fields[1]), fields[2].split("@")[0], fields[3]

def valid_lunch(lunch):
    return all(r[2] for r in lunch.ratings)
    


lunches = []
for line in lines:
    fields = line.split("\t")
    if fields[0] == "Timestamp":
        l = Lunch()
        l.prompt = re.findall(r"\(.*\)", fields[1])[0].strip("()")
        if l.prompt == "Honey and Thyme chicken and chorizo hotpot":
            l.prompt = "Honey Thyme Chicken Chorizo"
        lunches.append(l)
    else:
        rating = read_rating(fields)
        if rating:
            lunches[-1].ratings.append(rating)
    
valid_lunches = filter(valid_lunch, lunches)
all_raters = set([r[2] for lunch in valid_lunches for r in lunch.ratings if r[2]])


rows = []
for lunch in valid_lunches:
    lunch.make_dict()
    columns = dict((name, np.NAN) for name in all_raters)
    for r in lunch.ratings:
        columns[r[2]] = r[1]
    rows.append(columns)
lunch_names = [lunch.prompt for lunch in valid_lunches]
    
people_df = pd.DataFrame(rows, lunch_names)
lunches_df = people_df.transpose()

standardized_df = (people_df - people_df.mean())/people_df.std()
stand_freq = standardized_df.get(frequent_people)
std_lunches_df = standardized_df.transpose()

frequent_df = lunches_df[people_df.count() > 10].transpose()
frequent_people = list(frequent_df.columns)
```

### The important part - what do we like?

The first and the last place will not be a surprise anyone following daily lunch surveys. The big surprise here is the high placement of soup - and from (at least) two different vendors.

```python
lunch_counts = lunches_df.count().groupby(lambda x: x).sum()
frequent_lunches = list(lunch_counts[lunch_counts > 12].keys())
# frequent_lunches
avg_lunch_scores = people_df.groupby(lambda x: x).mean().transpose().mean()
fq_lunch_scores = avg_lunch_scores.get(frequent_lunches)
fq_lunch_scores.sort()

n = len(fq_lunch_scores)
scores = fq_lunch_scores.values
fig, ax = plt.subplots()
fig.set_size_inches(13, 7)
index = np.arange(n)
bar_width = 0.7

opacity = 0.4

rects1 = ax.bar(index, scores, bar_width)

ax.set_xlim(-bar_width, n + bar_width)
ax.set_ylim(1, 5)
ax.set_ylabel('ratings')
ax.set_title('lunches')
ax.set_title('avg rating per lunch')

# plt.xticks(index + bar_width, frequent_lunches)
ax.set_xticks(index + bar_width/2)
xtickNames = ax.set_xticklabels(fq_lunch_scores.keys())
plt.setp(xtickNames, rotation=90)
plt.show()
```

Creating this plot and all the others, I only used lunches/people with some minimum number of datapoints. In this case only lunches with more than 12 ratings made the cut. Also, in some cases I aggregated similar lunches into a single entry - hence "burgers".

It's hard to give this result a more precise interpretation though because some of us grade on a curve while others don't. ### This is all very fine, but why so monochromatic? Can we have more fun?

Let's dig deeper. Can we find any predictable structure in ratings? Lets start with the correlation matrix. Correlation matrix is a square showing how aligned people's ratings are. The highest possible correlation is 1 - means these raters rate in unison. The lowest is -1 meaning that they have exactly opposite tastes. Correlation 0 means no linear relationship. The figure below shows correlations as color - you can read the correlation between e.g. Nick and Tom by looking at the intersection of Tom's row with Nick's column (or the other way around - result will be the same). 

```python
# peeps = ['ahmed', 'chi', 'scott', 'andrewmw', 
#          'tessa','andy', 'nadbor','nick', 'cedric',
#          'tom', 'lucy','janno', 'victoria', 'michal.zlabek', 'andrew.warren-payne']
# with matplotlib.rc_context():
#     matplotlib.rc("figure", figsize=(14,10))
#     sm.graphics.plot_corr(frequent_df.get(peeps).corr(), xnames=peeps)
average_frittata_score = stand_freq.transpose()['frittata'].transpose().mean()
average_frittata_score.sort()

peeps = [name for name, score in average_frittata_score.iteritems() if score]

with matplotlib.rc_context():
    matplotlib.rc("figure", figsize=(14,10))
    sm.graphics.plot_corr(stand_freq.get(peeps).corr(), xnames=peeps)
```

The diagonal is all red - meaning correlation 1 - as should be expected - everyone is perfectly correlated with him/her-self. What else can we see? The square is mostly orange-yellowy which means that different raters tend to rate high the same lunches - no surprise there. Secondly, we can see kind of a reddish cluster in the upper left quarter and a smaller one in the lower in the lower right. These indicate a group of people with stronger than average correlation. Is there some deeper meaning to this? A clue: people in this figure are ordered according to how high they rated frittata. So the ones on in the upper left are frittata haters and on the other end of the diagonal sit frittata lovers. The clusters in this figure are rather vague but we can make them more apparent by taking only the clearest cases and reordering people.

```python
peeps = ['ahmed', 'chi', 'scott', 'andrewmw', 'roi','alex', 'nick',
         'cedric', 'tessa',
         'lucy','janno', 'victoria', 'michal.zlabek', 'andrew.warren-payne']
with matplotlib.rc_context():
    matplotlib.rc("figure", figsize=(14,10))
    sm.graphics.plot_corr(frequent_df.get(peeps).corr(), xnames=peeps)
```

Now the clusters are impossible to miss. Correlations inside cluster are very high (red) while  between clusters are low or negative (blue). These new clusters are also easier to interpret. Ahmed and Scott are card-carrying junkfood afficionados while Janno, Lucy and Victoria are known for their salad leanings.

A similar analysis can be performed on lunches instead of people. This time correlation between lunches means the degree to which these two lunches tend to get the same rating from the different raters.

```python
lun = ['hummus bros', "Karma Cans", "Hiba Express",
       'frittata', 'surf & turf wraps', 'Pure', 'quiche',
       'burgers', 'premium soup', 'soup',  
       'sushi', "Herman ze German", "LeCoq"]
# frequent_df.groupby(lambda x: x).mean().transpose().get(lun)
lunches_people_avg = frequent_df.groupby(lambda x: x).mean().transpose().get(lun)
# frequent_df.groupby(lambda x: x).mean().transpose().get(lun)

# print lunches_people_avg.corr()
with matplotlib.rc_context():
    matplotlib.rc("figure", figsize=(14,10))
    sm.graphics.plot_corr(lunches_people_avg.corr(), xnames=lun)
```

Not surprisingly, hummus, Karma Cans salads and Hiba Express form a highly correlated cluster.
